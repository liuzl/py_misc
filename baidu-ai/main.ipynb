```python
from aip import AipOcr

APP_ID = ''
API_KEY = ''
SECRET_KEY = ''

client = AipOcr(APP_ID, API_KEY, SECRET_KEY)

def get_file_content(filePath):
    with open(filePath, 'rb') as fp:
        return fp.read()
```

```python
image = get_file_content('example.jpg')
ret = client.passport(image)
```
