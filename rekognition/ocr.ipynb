```python
import boto3
from IPython.display import Image
client = boto3.client('rekognition', region_name='us-west-2')
```

```python
img = '0000547b3cff268a9d44b0b903acf5b9.jpg'
Image(img, width=500)
```

```python
resp = client.detect_text(Image={'Bytes':open(img,'rb').read()})
```

```python
for item in resp['TextDetections']:
    print(item['DetectedText'])
```
